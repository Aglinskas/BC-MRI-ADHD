```python
def pad_subID(val):
    if np.isnan(val):
            val_str = val
    else:
        val_str = str(int(val))
        if len(val_str)==5:
            val_str = 'sub-00'+val_str
        if len(val_str)==7:
            val_str = 'sub-'+val_str
    return val_str

df['subID'] = [pad_subID(val) for val in df['ScanDir ID'].values]
df.to_csv('../Data/df_440_normed_padded.csv')
```
